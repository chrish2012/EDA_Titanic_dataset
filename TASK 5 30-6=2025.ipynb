# step-3: Data Visualization (using Seaborn & Matplotlib)

```python
#Distribution plots (Univariate Analysis)

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore")
sns.histplot(data['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

# observations 

### 1 Most passengers were between 20 to 40 years old, with a clear peak near 30 years.

### 2 The distribution is right-skewed, meaning few passengers were older (above 60).

### 3 Some children under 10 also boarded the Titanic, showing diverse age groups.

```python
# Boxplot (Outliers check)

sns.boxplot(x='Pclass', y='Age', data=data)
plt.show()
```

# observation

### 1st Class passengers generally had higher ages, with many above 40.

###  2nd and 3rd Class passengers were younger on average, with more extreme outliers.

###  This suggests that older, possibly wealthier individuals preferred higher classes.

```python
# countplot
sns.countplot(x='Survived',data=data)
sns.countplot(x='Sex',hue='Survived',data=data)
```

# Observation

### A significantly higher number of females survived compared to males.

### Males had a much lower survival rate, with more male passengers overall.

### This confirms the famous Titanic policy of "women and children first" during evacuation.

```python
# Heatmap (Correlation Matrix)
columns = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp',
       'Parch', 'Fare']

# Create the correlation matrix for the selected columns
corr_matrix = data[columns].corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Matrix of Financial Data")
plt.show()
```

# Observation:

### Fare and Pclass show a strong negative correlation (-0.55): higher-class passengers (Pclass=1) paid more fare.

### Parch and SibSp have positive correlation (0.41): passengers traveling with parents often also had siblings/spouses onboard.

### Survived shows weak positive correlation with Fare (0.26) and weak negative correlation with Pclass (-0.34), indicating passengers who paid more and were in higher classes had better survival chances.

### Age has low correlation with survival, suggesting age alone didnâ€™t strongly affect outcome.



```python
# Pairplot (Multivariate Analysis)
import warnings
warnings.simplefilter("ignore")
sns.pairplot(data[['Age', 'Fare', 'Pclass', 'Survived']])
```

# Observation:

### Fare vs Survived: Passengers who paid higher fare had a higher chance of survival, with many survivors clustered at the upper range.

### Age distribution is dense between 20â€“40 years, with survivors spread across all ages.

### Pclass and Fare show visible clustering: 1st class passengers mostly paid more.

### The survival variable forms distinct clusters in multiple pairwise combinations, confirming weak to moderate influence of variables like Fare, Pclass, and Sex on survival.
